```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X = np.array([
    [15, 39], [15, 81], [16, 6], [16, 77], [17, 40],
    [17, 76], [18, 6], [18, 94], [19, 3], [19, 72],
    [20, 14], [20, 99], [21, 15], [21, 77], [22, 13],
    [22, 79], [23, 35], [23, 66], [24, 29], [24, 98],
    [25, 35], [25, 73], [26, 5], [26, 73], [27, 14],
    [27, 82], [28, 32], [28, 61], [29, 31], [29, 87]
])
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']
for i in range(3):
    cluster_points = X[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                s=100, c=colors[i], label=f'Cluster {i+1}')
plt.scatter(centroids[:, 0], centroids[:, 1],
            s=200, c='yellow', marker='X', label='Centroids')
plt.title("Customer Clusters based on Purchase History")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1â€“100)")
plt.legend()
plt.grid(True)
plt.show()
```
